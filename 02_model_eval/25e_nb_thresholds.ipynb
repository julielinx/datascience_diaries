# Entry 25 notebook - Setting thresholds - MNIST

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate, cross_val_predict
from sklearn.compose import ColumnTransformer, make_column_transformer, make_column_selector
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import SGDClassifier

from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

from sklearn.datasets import load_digits, fetch_openml
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], 'b--', label='precision')
    plt.plot(thresholds, recalls[:-1], 'g-', label='recall')
    plt.xlabel('Threshold')
    plt.legend(loc='center right')
    plt.grid()
    plt.ylim([0, 1])
    
def plot_precision_vs_recall(precisions, recalls, thresholds):
    close_zero = np.argmin(np.abs(thresholds))
    plt.plot(precisions[close_zero], recalls[close_zero], 'o', markersize=10,
            label='threshold zero', fillstyle="none", c='k', mew=2)
    plt.plot(recalls, precisions, "b-", linewidth=2, label='precision recall curve')
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.legend(loc='best')
    plt.grid(True)
    
def plt_roc_curve(fpr, tpr, thresholds, label='ROC curve'):
    close_default = np.argmin(np.abs(thresholds))
    plt.plot(fpr[close_default], tpr[close_default], 'o', markersize=10,
            label='default threshold 0.5', fillstyle="none", c='k', mew=2)
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0,1], [0,1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
```

## Make dataset and fit model

```python
mnist = fetch_openml('mnist_784', version=1)
X = mnist.data
y = mnist.target

X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]

y_train_5 = (y_train == '5')
y_test_5 = (y_test == '5')
```

```python
print(X.shape)
print(y.shape)
```

```python
pipe = make_pipeline(SimpleImputer(strategy='median'), StandardScaler(),
                     SGDClassifier(random_state=42))
pipe.fit(X_train, y_train_5)
```

```python
cross_validate(pipe, X_train, y_train_5, cv=3, scoring=['precision', 'average_precision', 'recall'])
```

### Precision vs Recall

```python
y_scores = cross_val_predict(pipe, X_train, y_train_5, cv=3, method='decision_function')
```

```python
precisions, recalls, pr_thresholds = precision_recall_curve(y_train_5, y_scores)
plot_precision_recall_vs_threshold(precisions, recalls, pr_thresholds)
plt.xlim([-500, 500])
```

```python
plot_precision_vs_recall(precisions, recalls, pr_thresholds)
```

### ROC Curve and ROC AUC

```python
fpr, tpr, roc_thresholds = roc_curve(y_train_5, y_scores)
plt_roc_curve(fpr, tpr, roc_thresholds)
```

```python
roc_auc_score(y_train_5, y_scores)
```
