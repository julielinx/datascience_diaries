## Figure Factory

The explaination of figure factories is extremely short (2 sentences), but seems to be a way to do charts for specialized domains.

```python
import numpy as np
import plotly.figure_factory as ff

x1, y1 = np.meshgrid(np.arange(0, 2, .2), np.arange(0, 2, .2))
u1 = np.cos(x1)*y1
v1 = np.sin(x1)*y1

fig = ff.create_quiver(x1, y1, u1, v1)

fig
```
