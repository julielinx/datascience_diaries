## Modeling 3D linear functions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from math import pi
```

```python
planet_df = pd.read_excel('../data/planets_moons.xlsx')
p_cols = ['name', 'mass_1024kg', 'diameter_km','mean_radius_km', 'density_kg_m3', 'gravity_m_s2', 'escape_vel_km_s']
planet_df = planet_df[p_cols]
planet_df['volume'] = round(planet_df['mean_radius_km'] ** 3 * (4/3) * pi, 4)
planet_df
```

- direct measures
  - mass (M)
  - diameter (D) - 2r
  - radius (r) - $\frac{1}{2} D$
  - gravatational constant (G)
- density = mass / volume
- volume = $\frac{4}{3} \pi r^3$
- gravity = $\frac{GM}{r^2}$
- escape velocity = $\sqrt{\frac{2GM}{r}}$

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(planet_df['mass_1024kg'], planet_df['volume'], planet_df['density_kg_m3'])

ax.set_xlabel('mass')
ax.set_ylabel('volume')
ax.set_zlabel('density')
ax.set_title('Density')
```
