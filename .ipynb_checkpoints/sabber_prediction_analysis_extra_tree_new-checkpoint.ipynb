### Impute_missing_value

```python
df_train = df_train.replace(r'^\s*$', np.nan, regex=True)
df_train.fillna(value = -1, inplace = True)

df_test = df_test.replace(r'^\s*$', np.nan, regex=True)
df_test.fillna(value = -1, inplace = True)
```

### Remove_high_missing_features

```python
number_rows = df_train.shape[0]
df_missing = df_train.isin([-1]).sum(axis = 0).to_frame('miss_count').reset_index().rename(columns={"index": "features"})
df_missing['miss_percent'] = df_missing['miss_count'].map(lambda x: (x*100)/number_rows)
high_missing_features = df_missing[df_missing.miss_percent >= 90]['features'].tolist()
if len(high_missing_features) > 0:
    df_train.drop(high_missing_features, axis = 1, inplace  = True)
    df_test.drop(high_missing_features, axis = 1, inplace  = True)
```

```python
df_train.shape, df_test.shape
```

### Remove duplicate_rows

```python
df_train = df_train.drop_duplicates()
df_test = df_test.drop_duplicates()
```

```python
df_train.shape, df_test.shape
```

### Create date time features

```python
date_time_column = 'sales_offer_date'
df_train[date_time_column] = pd.to_datetime(df_train[date_time_column])
df_train['datetime_day_of_the_week'] = df_train[date_time_column].dt.dayofweek
df_train['datetime_hour_of_the_day'] = df_train[date_time_column].dt.hour
df_train['datetime_day_of_the_month'] = df_train[date_time_column].dt.day
df_train_date_col = df_train[date_time_column]
df_train.drop([date_time_column], axis = 1, inplace = True)

df_test[date_time_column] = pd.to_datetime(df_test[date_time_column])
df_test['datetime_day_of_the_week'] = df_test[date_time_column].dt.dayofweek
df_test['datetime_hour_of_the_day'] = df_test[date_time_column].dt.hour
df_test['datetime_day_of_the_month'] = df_test[date_time_column].dt.day
df_test_date_col = df_test[date_time_column]
df_test.drop([date_time_column], axis = 1, inplace = True)
```

```python
df_train.shape, df_test.shape
```

### encode_categorical_features

```python
### Find the features to be encoded
encoded_features = df_train.select_dtypes('O').columns.tolist()

### Target encoding starts here
encoder = ce.OrdinalEncoder(cols = encoded_features)
encoder.fit(df_train, y = df_train[target].values, handle_missing = -1)

df_train = encoder.transform(df_train)
df_test =  encoder.transform(df_test)
encoder = encoder

new_names = [(i, 'categorical_' + i) for i in encoded_features]
df_train.rename(columns = dict(new_names), inplace=True)
df_test.rename(columns = dict(new_names), inplace=True)
```

```python
df_train.shape, df_test.shape
```

### remove_zero_std_features

```python
df = pd.DataFrame(df_train.std(axis = 0).reset_index()\
     .rename(columns={"index": "features", 0: 'std'}))
zero_std_features = df[df['std'] <= 0.0]['features'].tolist()

if len(zero_std_features) > 0:
    df_train.drop(zero_std_features, axis = 1, inplace = True)
    df_test.drop(zero_std_features, axis = 1, inplace = True)
```

```python
df_train.shape, df_test.shape
```

### remove_duplicate_features

```python
groups = df_train.columns.to_series().groupby(df_train.dtypes).groups
dup_features = []
for t, v in groups.items():
    cs = df_train[v].columns
    vs = df_train[v]
    lcs = len(cs)
    for i in range(lcs):
        ia = vs.iloc[:,i].values
        for j in range(i+1, lcs):
            ja = vs.iloc[:,j].values
            if np.array_equal(ia, ja):
                dup_features.append(cs[i])
                break
                        
if len(dup_features) > 0:
    df_train.drop(dup_features, axis = 1, inplace = True)
    df_test.drop(dup_features, axis = 1, inplace = True)
```

### Balance train data

```python
def balance_data(xtrain, ytrain, technique):
    if technique == 'SMOTE':
        sampler = SMOTE(n_jobs = 16)
    elif technique == 'UnderSampling':
        sampler = RandomUnderSampler()
    elif technique == 'OverSampling':
        sampler = RandomOverSampler()
    elif technique == 'KMeansSMOTE':
        sampler = KMeansSMOTE(k_neighbors = 2, n_jobs = -1)
    elif technique == 'BorderlineSMOTE':
        sampler = BorderlineSMOTE(n_jobs = 16)
    else:
        return xtrain, ytrain
    
    xtrain, ytrain = sampler.fit_sample(X = xtrain, y = ytrain)
    return xtrain, ytrain
```

```python
xtrain_bal, ytrain_bal = balance_data(xtrain, ytrain, 'OverSampling')
```

```python
np.shape(xtrain_bal)
```

### Hyperparameter tuning

```python
from hyperopt import hp
from hyperopt.pyll import scope
from hyperopt.pyll.stochastic import sample
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials, space_eval
```

```python
hyperparameters = {
   'n_estimators': hp.choice('n_estimators', np.arange(5, 700, dtype = int)),
   'max_features': hp.choice('max_features', np.arange(1, 20, dtype = int)),
   'max_depth': hp.choice('max_depth', np.arange(5, 200, dtype = int)),
   'min_samples_split': hp.choice('min_samples_split', np.arange(2, 50, dtype = int)),
   'min_samples_leaf': hp.choice('min_samples_leaf', np.arange(1, 50, dtype = int)),
   'bootstrap': hp.choice('bootstrap', [True, False]),
    'class_weight': {'0': 1, '1': hp.choice('1', np.arange(1, 20, dtype = int))},
   'random_state': 2020,
    'n_jobs': -1
}
```

```python
def objective_function(param):
    param['class_weight'] = {0: param['class_weight']['0'], 1: param['class_weight']['1']}
    clf = ExtraTreesClassifier(**param)
    clf.fit(xtrain_bal, ytrain_bal)
    pred = clf.predict_proba(xtest)[:, 1]
    score = roc_auc_score(ytest, pred)
    return {'loss': -score, 'status': STATUS_OK}
```

```python
trials = Trials()
hyper_space = {}
best_param = fmin(objective_function, 
                  hyperparameters, 
                  algo = tpe.suggest, 
                  max_evals = 45,
                  trials = trials,
                  verbose = 100,
                  show_progressbar = True)

best_param = space_eval(hyperparameters, best_param)
```
