```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
planets = pd.read_excel('data/planets_moons.xlsx', index_col=0)
```

```python
def convert_categoricals(df):
    cat_vars = df.select_dtypes('object').columns.tolist()
    converted = df.copy()
    converted[cat_vars] = df[cat_vars].astype('category')
    for cat in cat_vars:
        converted[cat] = converted[cat].cat.codes
    return converted
```

```python
def feature_corr(df, target, corr_type='spearman'):
    target_corr = pd.DataFrame(df.corrwith(df[target], axis=0, method=corr_type).reset_index()).rename(columns={0:corr_type})
    target_corr['abs'] = target_corr[corr_type].abs()
    top_features = target_corr[target_corr['abs'] >= 0.5].sort_values('abs', ascending=False)['index'].tolist()
    df_top = df[top_features]
    feature_corr = df_top.corr(method=corr_type)
    return df_top, feature_corr
```

```python
planets_num = convert_categoricals(planets)
df_top, feature_corr = feature_corr(planets_num, 'atmospheric_mass_kg')
df_top
```

### Automatically Drop Collinear Features

Sabber devised a way to automatically drop correlated features.

A correlation matrix includes redundant information. A feature is 100% correlated to itself, which creates a diagonal line of ones from the top left to the bottom right. The values above this diagnoal are a mirror of the values below it. This can be seen in the subset of the correlation of planet features below.

```python
feature_corr.iloc[-5:, -5:]
```

Sabber used this diagonal divider to look for correlations over 0.9. By starting at the first row and only looking at columns before the feature's correlation with itself, Sabber ensured that the first feature of a correlated pair will be retained, but the second, and any others that follow, will not.

The conditional statement of the function is a little misleading. The sum() portion of the if statement is really only important if it runs. If there is a sum, that means there is at least 1 correlation over 0.9 with one of the features in the row above it. A different function could just as easily be used, like count().

```python
def feature_coll(correlation_matrix, df_top):
    collinear_features = set()
    for i in range(len(correlation_matrix.columns)):
        sliced_matrix = correlation_matrix.iloc[i, :i]
        if sum(sliced_matrix[abs(sliced_matrix) > 0.9]):
            colname = correlation_matrix.columns[i]
            collinear_features.add(colname)
    collinear_features = list(collinear_features)
    print('Collinear features:', collinear_features)
    df = df_top.drop(collinear_features, axis=1)
    return df

noncoll_df = feature_coll(df_top.drop('atmospheric_mass_kg', axis=1).corr(method='spearman'), df_top[sorted_features].drop('atmospheric_mass_kg', axis=1))
noncoll_df
```
