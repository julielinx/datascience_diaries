## Mathematics

- [ETEC 150 - Applied Technical Math](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115142)
    - Description: Using algebra and trigonometry to solve problems in electronics and manufacturing. Also covers use of scientific calculator, binary, and hexadecimal.
    - Prerequisites: MATH 087 (Essentials of Intermediate Algebra)
- [ITAD 315 - Discrete Mathematics for Developers](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115449)
    - Description: Students are introduced to logic and proof, structures and algorithms, and number theory. Focus is also given to induction and recursion and counting and discrete probability to provide the essential foundation and framework for software development.
    - Prerequisites: MATH& 141 (Precalculus)
- [MATH 151: Calculus I](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114751)
    - Description: Limits, continuity, derivatives, and applications. Differentiation of algebraic, exponential, logarithmic, trigonometric, and inverse-trigonometric functions.
    - Prerequisites: MATH& 142 (Precalculus II)
- [MATH 152: Calculus II](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114752)
    - Description: The integral and its applications. Integration involving algebraic, exponential, logarithmic, trigonometric, and inverse-trigonometric functions
    - Prerequisites: MATH& 151
- [MATH 153: Calculus III](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114753)
    - Description: Sequences, series, power series, vectors, vector-valued functions, lines and planes in 3D, calculus with parametric equations and polar coordinates, and partial derivatives.
    - Prerequisites: MATH& 152
- [MATH& 254 - Calculus IV](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114754)
    - Description: Partial differentiation, multiple integration, vector fields, line and surface integrals, Green’s Theorem, Stokes’ Theorem, Divergence Theorem, and applications.
    - Prerequisites: MATH& 153
- [MATH 271 - Differential Equations](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114743)
    - Description: First- and second-order differential equations and their applications, Laplace transforms, and series solutions of ordinary differential equations.
    - Prerequisites: MATH& 153
- [MATH 272 - Linear Algebra](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114744)
    - Description: Matrices, systems of equations, vector spaces, linear transformations, and eigenvalues.
    - Prerequisites: MATH& 153

## Computer Science

- [CS 115 - Introduction to Programming](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114401)
    - Description: An introductory course in programming using VB.NET. No previous programming experience is expected. Topics include designing, creating and debugging interactive, event-driven programs with a graphical user interface and developing problem solving skills.
    - Prerequisites: MATH 090 or MATH 097 or ETEC 150
- [CS& 141 - Computer Science I Java](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114425)
    - Description: Introduction to Java programming. Topics include basic Java syntax, data types, control structures, methods, object representation using classes, graphics and arrays, all within a framework of general object oriented programming principles.
    - Prerequisites: CS 115
- [CS 142 - Computer Science II Java](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114406)
    - Description: Intermediate Java programming. Topics include algorithm development, searching/sorting, complexity/efficiency, recursion, user interface design, class relationships including composition and inheritance and an introduction to abstract data types.
    - Prerequisites: CS& 141
- [CS 143 - Computer Science III Java](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114407)
    - Description: Java Data Structures. Topics include data structures such as lists, stacks, queues and various binary trees, inheritance and interfaces, using standard collection classes and algorithms and Generics.
    - Prerequisites: CS 142
- [CS 125 - Linux and UNIX I](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114402)
    - Description: First of a two-quarter survey of Linux/UNIX operations. Topics include base Linux commands; combining commands to create utilities; managing files, software and processes; creating partitions and Logical Volumes; editing text; managing users and groups; extended attributes and basic networking.
    - Prerequisites: None
- [CS 126 - Linux and UNIX II](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114403)
    - Description: Second of a two-quarter survey of Linux/UNIX operations. Topics include Boot process, process management, RPM, creating/optimizing RAID, encryption, managing user access, configuring dynamic/static networks, Bash scripting, Apache secure virtual hosts, Samba/NFS, Postfix mail server, SSH and NX
    - Prerequisites: CS 125
- [CS 225 - UNIX Shell Programming](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114418)
    - Description: An intermediate course that extends previous experience with the Bourne/bash shells to program scripts used to automate system administrative tasks. Topics include environment/user defined variables, branches, loops, menus, user interaction, and functions
    - Prerequisites: CS 125
- [CS 290 - Intro to Software Testing and Quality Assurance](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114422)
    - Description: Covers the fundamental concepts and techniques of software testing and quality assurance. Topics include goals of testing and quality assurance, classification of bugs, testing categories and techniques, test design, metrics and complexity.
    - Prerequisites: CS 115
    
## Computer Information Systems

- [CIS 233 - Systems Analysis](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114300)
    - Description: A practical approach to real world systems analysis and design. Includes the systems development life cycle, agile development, structured methodologies, and project planning. A case study project is analyzed, requirements are written, and a systems design specification document is prepared.
    - Prerequisites: CIS 102
- [CIS 234 - Systems Design and Development](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114301)
    - Description: Provides hands-on experience in systems design and implementation using a prototype approach. The case study is designed and tested using application development software. User manuals are written.
    - Prerequisites: CIS 233
- [CIS 241 - Web Development I](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114302)
    - Description: Website development using HTML5, approached from a source code perspective. Covers tags, forms, linked objects, CSS3, frames, tables, and introduction to the use of scripting. Students build multi-page websites using GUI HTML editors, Git and GitHub.
    - Prerequisites: CIS 100
- [CIS 242 - Web Development II: JavaScript and jQuery](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114303)
    - Description: Students will explore the topics of embedding, inline and external scripts, functions, form validation, loops, conditional statements, strings, numbers, DHTML, and an introduction to JavaScript Frameworks. Introduction to creating Node.js websites. Students build multi-page websites using JavaScript technologies.
    - Prerequisites: CIS 241
- [CIS 243 - Web Development III: Introduction to PHP](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114304)
    - Description: Topics include server-side scripting for web pages using PHP. Students will learn fundamentals of PHP scripting including functions, logical structure, and database connectivity used in creating websites. Other topics include Object-Oriented PHP, MySQL, Ruby on Rail, and Django Python web frameworks.
    - Prerequisites: CIS 242
- [CIS 246 - Web Dev VI: JavaScript Frameworks](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115047)
    - Description: Students will investigate and use the advanced techniques available in JavaScript frameworks (libraries) such as JQuery, AJAX, MooTools, AngularJS, ReactJS, and Prototype.
    - Prerequisites: CIS 242
- [CIS 250 - Database Theory and Design](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114306)
    - Description: Designed to recognize data as a business resource. Database models are discussed from both a developer’s and a user’s viewpoint. Topics include: conceptual and logical models, cloud computing, big data, data warehouses, business intelligence, and data normalization through case studies.
    - Prerequisites: CIS 102
- [CIS 253 - Application and Cloud Development Tools](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=114308)
    - Description: An introduction to application development using MS-Access and WordPress, Drupal, Odoo, SugarCRM, Magento Web frameworks. Includes development of tables, forms, queries, reports, macros, and menus needed to support the application. Includes application development in the Cloud platforms, such as Microsoft Azure or Amazon Web Services (AWS).
    - Prerequisites: CIS 250 and CS 115

## Engineering Technology

- [ETEC 150 - Applied Technical Math](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115142)
    - Description: Using algebra and trigonometry to solve problems in electronics and manufacturing. Also covers use of scientific calculator, binary, and hexadecimal.
    - Prerequisites: MATH 087 (Essentials of Intermediate Algebra)
- [ETEC 140 - Basic Electricity and Electronics](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115559)
    - Description: Basics of AC and DC circuits. Includes resistors, capacitors, RC circuits, inductors, filters, diodes,transistors, and operational amplifiers. Testing and troubleshooting circuits where voltage changes over time. Practice using DMM and oscilloscope.
    - Prerequisites: ETEC 150
- [ETEC 141 - Digital and Microprocessor Electronics](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115560)
    - Description: Introduction to digital and microprocessor circuits. Covers operation of basic logic gates, flip-flops, decoders, displays, counters, and microprocessor circuits. Lectures and hands-on exercises to learn how to analyze, test, and troubleshoot digital and microprocessor circuits.
    - Prerequisites: ETEC 140
- [ETEC 142 - Automation Programming](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115561)
    - Description: Through the use of functional testing with an oscilloscope, students will learn to troubleshoot a robotic system.Students will also program and build a robot system using the Arduino prototyping platform.
    - Prerequisites: ETEC 140
- [ETEC 143 - Automation Programming and PLCs](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115562)
    - Description: Students will learn the basics of Programmable Logic Controllers (PLCs) used to operate automated systems and how to program and communicate with the systems.
    - Prerequisites: ETEC 141
- [ETEC 242 - Advanced Analog and Digital Electronics](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115563)
    - Description: Advanced training in analog and digital electronics. Covers analysis, functional testing, and systematic troubleshooting of analog circuits and digital circuits designed with op-amp amplifiers and comparators, mux, demux chips, shift registers, memory circuits, latches, and buffers.Includes basic soldering techniques.
    - Prerequisites: ETEC 141
- [ETEC 245 - Robotics in Automation](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115578)
    - Description: Through the use of projects and hands-on exercises, students will learn to program a small scale robot to perform basic navigation under program control, with various sensor inputs, using feedback, and control techniques.
    - Prerequisites: ETEC 140


## Robotics AI

- [ROBAI 370 - Advanced Computer Vision and Sensors](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115623)
    - Description: Robotics sensing systems give robots the ability to see, touch, hear, and move and usealgorithms to provide feedback to the robots and control their behavior. This course gives an overview of the current technologies and trends in robotics sensing. Students learn to analyze sensing requirements of various robotics applications, and design basic sensing systems based in these specifications.
    - Prerequisites: ETEC 245
- [ROBAI 401 - Language and Speech Technology](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115625)
    - Description: The course offers a practical understanding of how human speech can be processed by computers and robots. Students build speech recognition, speech synthesis, and spoken dialog systems. Topics include algorithms, techniques, and limitations of state-of-the-art speech systems.
    - Prerequisites: CIS 233 and ROBAI 380
- [ROBAI 410 - Autonomous Systems](https://catalog.edmonds.edu/preview_course_nopop.php?catoid=56&coid=115627)
    - Description: This course focuses on key components of an autonomous robotic system and explores robot interaction with the environment, human-robot interactions, and multi-robot cooperation. Students learn to integrate data from multiple sensor inputs to support robot navigation, and develop models to control robot behavior, decision making, and autonomy.
    - Prerequisites: ROBAI 370 and ROBAI 380
