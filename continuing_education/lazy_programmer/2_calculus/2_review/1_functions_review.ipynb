# Functions Review

## Equations/Functions from the lecture

### Random

- $\text{height} = -\frac{1}{2}gt^2 + v_{0}t + h_{0}$
- $\text{area} = \pi r^2$

## Basic

- $\text{line} = mx + b$
- $\text{polynomial}_{a} = ax^3 + bx^2 + cx + d$
- $\text{root_fx}_{a} = ax^\frac{1}{3} + bx^\frac{1}{2} + c$
- $\text{exponential_fx}_{a} = 2^x$
- $\text{log}_{10} = log x$
- $\text{log}_{e} = ln x$

### Trigonometric

- $\text{y} = sin x = \frac{opposite}{hypotenuse}$
- $\text{y} = cos x = \frac{adjacent}{hypotenuse}$
- $\text{y} = tan x = \frac{opposite}{adjacent} = \frac{sin x}{cos x}$
- $\text{y} = \frac{sin x}{x} = sin c(x)$

### Hyperbolic

- $\text{y} = sin h(x) = \frac{e^x - e^{-x}}{2}$
- $\text{y} = cos h(x) = \frac{e^x + e^{-x}}{2}$
- $\text{y} = tan h(x) = \frac{sin h(x)}{cos h(x)} = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

## Line: mx + b

```python
x = np.linspace(-5, 5, 100)
x
```

```python
# slope = 0.5, y-intercept=1
y = 0.5*x+1
y
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, c='steelblue');
```

```python
# slope = 2, y-intercept=-3
y = 2*x - 3
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, c='cyan')
plt.axis('scaled');
```

```python
# slope = -.5, y-intercept = 7
y = -0.5*x +7
plt.plot(x,y, c='m')
plt.axis('equal')
```

```python
# slope = -.5, y-intercept = 7
y = -0.5*x +7
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x,y, c='cornflowerblue');
```

## Quadratic ~= polynomial: $-4.9x^2 + 2x + 1$

```python
x = np.linspace(-2, 2, 100)
y = -4.9*x**2 + 2*x + 1
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, c='firebrick');
```

## Root Function

```python
x = np.linspace(0, 10, 100)
y = -0.5 * x**(1/3) + 0.2 * x**(1/2) + 1

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, c='darkorange')
```

## Exponential

```python
x = np.linspace(-5, 5, 100)
y = np.exp(x) # e^x

plt.axvline(x=0, c='k')
plt.axhline(y=0, c='k')
plt.grid()
plt.plot(x, y, c='darkolivegreen')
```

```python
x = np.linspace(-5, 5, 100)
y = np.exp(x) # e^x
y2 = x**2

plt.axvline(x=0, c='k')
plt.axhline(y=0, c='k')
plt.grid()
plt.plot(x, y, c='royalblue', label='exponential')
plt.plot(x, y2, c='firebrick', label='squared')
plt.legend();
```

## Logarithmic

```python
x = np.linspace(0.0000001, 10, 200)
y = np.log(x) # in np library, this is base e: ln(x)

plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.grid()
plt.plot(x, y, c='darkred');
```

```python
# This isn't what the tangent line actually looks like
# The line goes toward infinity at pi/.
# The computer tries to draw these lines in finite space,
# making the tangent line look like a HR
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.tan(x) 

plt.figure(figsize=(15,5))
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.grid()
plt.plot(x, y);
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100000)
y = np.tan(x) 

plt.figure(figsize=(15,5))
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.grid()
plt.plot(x, y);
```

```python
# This is what it should look like
x = np.linspace(-2 * np.pi, 2 * np.pi, 100000)
y = np.tan(x) 

plt.figure(figsize=(15,5))
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.grid()
plt.plot(x, y)
plt.ylim(-60, 60);
```

## Hyperbolic

```python
x = np.linspace(-5, 5, 100)
y = np.sinh(x)

plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.grid()
plt.plot(x, y);
```

```python
# Are sinh and cosh the same in the range of positive x?
x = np.linspace(-5, 5, 100)
y = np.sinh(x)
y2 = np.cosh(x)

plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.grid()
plt.plot(x, y, c='royalblue', label='sinh')
plt.plot(x, y2, c='firebrick', label='cosh');
```

```python
# This is the sigmoid
# It's what's used in logistic regression
x = np.linspace(-5, 5, 100)
y = np.tanh(x)

plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.grid()
plt.plot(x, y);
```
