## Definitions and Equations

$f`(a)$ or $f^{\prime}(a)$ = f prime a

The derivative of a function $f$ at a number $a$ denoted by $f^{\prime}(a)$ is:

$\quad f^{\prime}(a) = \lim\limits_{h \to 0} \frac{f(a+h) - f(a)}{h}$\
or\
$\quad f^{\prime}(a) = \lim\limits_{x \to a} = \frac{f(x) - f(a)}{x - a}$

These equations hold true when "a number $a$" is replaced with a variable $x$.

$\quad f^{\prime}(x) = \lim\limits_{h \to 0} \frac{f(x+h) - f(x)}{h}$

$f$ is only differentiable at any given value $a$ if $f$ is continuous at $a$. So the derivative can be determined if the value is the same approaching from the left ($-$) and from the right ($+$):

$\quad f^{\prime}(a) = \lim\limits_{h^- \to 0} \frac{f(a+h) - f(a)}{h}$ \
and \
$\quad f^{\prime}(a) = \lim\limits_{h^+ \to 0} \frac{f(a+h) - f(a)}{h}$

Types of functions that are not differentiable:

- Cusp
- Discontinuity
- Vertical tangent

### Corner

A corner has two different tangents in any direction.

```python
x = np.linspace(-10, 10, 1000)
y = abs(x)

plt.figure(figsize=(14, 4))
plt.plot(x, y)
plt.grid()
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

```python
x_neg = np.linspace(-10, 0, 100)
y_neg = np.ones(len(x_neg))
x_pos = np.linspace(0, 10, 100)
y_pos = 2 * x_pos + 1
x = np.concatenate([x_neg, x_pos])
y = np.concatenate([y_neg, y_pos])

plt.figure(figsize=(14, 4))
plt.plot(x, y)
plt.grid()
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

### Discontinuity

```python
x_rt = np.linspace(0, 20, 100)
y_rt = ((x_rt - 4)**(2/3) - 3) * -1
y_lt = y_rt[::-1] - 3
x_lt = (x_rt[::-1] * -1) + 8.1
x = np.concatenate([x_rt, x_lt])
y = np.concatenate([y_rt, y_lt])

plt.figure(figsize=(14, 4))
plt.plot(x, y)
plt.grid()
plt.xlim(-12, 20)
plt.ylim(-3, 4)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

### Vertical tangent

At point (-5, 3) the slope is undefined (slope 0 would be a horizontal line. A vertical line would have a slope of $\infty$ or $-\infty$, which is officially not a number).

```python
x = np.linspace(-20, 10, 100)
y = np.cbrt(x + 5) + 3

# plt.figure(figsize=(14, 4))
plt.plot(x, y)
plt.grid()
# plt.xlim(-12, 20)
# plt.ylim(-3, 4)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

## ReLU

$\text{ReLU}(x) = \max(0, x)$

This means any negative number will return a max value of 0 and any positive number will return itself.

$\text{ReLU}(x) = \begin{cases}
    x \quad \text{if } x \ge 0 \\
    0 \quad \text{if } x < 0
  \end{cases}$

This results in the below graph:

```python
x_neg = np.linspace(-10, 0, 100)
y_neg = np.zeros(len(x_neg))
x_pos = np.linspace(0, 10, 100)
y_pos = x_pos
x = np.concatenate([x_neg, x_pos])
y = np.concatenate([y_neg, y_pos])

plt.figure(figsize=(14, 4))
plt.plot(x, y)
plt.grid();
```

# 2. More on Tangent Lines, Derivative Checking

In bringing the approximation of the tangent closer and closer to the tangent (closing the distance between $x_2$ and $x_1$), the change approaches becoming a single point.\
$\Delta x \to dx = \text{infinitesimal distance}$

### Checking derivatives

The derivative may be difficult to compute. In these instances, you can check your answer by using very small or very large values depending on what you're trying to calculate.

For example:

$\quad \lim\limits_{x \to 0} \frac{\sin x}{x} \simeq \frac{\sin(0.00001)}{0.00001}$ \
or \
$\quad \lim\limits_{x \to \infty} \frac{\sin x}{x} \simeq \frac{\sin(1000000000)}{1000000000}$

This gives an equation like:

$\quad \lim\limits_{x \to 0} \frac{f(a + h) - f(a)}{h} \simeq \frac{f(a + 0.00001) - f(a)}{0.00001}$

This method works even when you don't know $f$. You can just call the function for the result, then plug in the numbers. This comes in handy when working with neural networks or optimization algorithms.

### Approximating from the right

```python
x = np.linspace(-3, 2, 100)
y = np.exp(x)

plt.figure(figsize=(18, 4))
plt.subplot(1, 3, 1)
plt.plot(x, y, '-bo', markevery=[85])
plt.scatter(x[65], y[65], s=60, c='firebrick')
plt.plot(x[[65, 85]], y[[65, 85]])
plt.grid()
plt.subplot(1, 3, 2)
plt.plot(x, y, '-go', markevery=[77])
plt.scatter(x[65], y[65], s=60, c='firebrick')
plt.plot(x[[65, 77]], y[[65, 77]])
plt.grid()
plt.subplot(1, 3, 3)
plt.plot(x, y, '-co', markevery=[70])
plt.scatter(x[65], y[65], s=60, c='firebrick')
plt.plot(x[[65, 70]], y[[65, 70]])
plt.grid()
```
