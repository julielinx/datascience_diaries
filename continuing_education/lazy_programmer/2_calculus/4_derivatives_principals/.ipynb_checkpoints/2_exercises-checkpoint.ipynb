# $f(x) = x^2$

Find: $f \prime (x)$

$\lim\limits_{x \to 0} x^2 \simeq \lim\limits_{h \to 0} \frac{f(x+h) - f(x)}{h}$

$\lim\limits_{h \to 0} \frac{(x+h)^2 - x^2}{h}$

$\lim\limits_{h \to 0} \frac{x^2 + 2xh +h^2 - x^2}{h}$

$\lim\limits_{h \to 0} \frac{2xh +h^2}{h}$

$\lim\limits_{h \to 0} 2x + h$

$h = 0 \simeq 2x + 0 = 2x$

$\frac{d}{dx}(x^2) = 2x$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
y = x**2

plt.plot(x, y)
plt.grid()
plt.scatter(0, 0, c='k', s=50);
```

```python
# f(x) = x^2
# f'(x) = 2x

h = 1e-10
x = 5

print('Estimate:', ((x+h)**2 - x**2) / h)
print('True:', 2*x)
print('\n')
print('Estimate:', ((x+h)**2 - (x-h)**2) / (2*h))
print('True:', 2*x)
```

```python
x = np.linspace(-10, 10, 2001)
y = x**2
yp = 10*x - 25

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.grid()
plt.scatter(x[np.where(x==5)], y[np.where(x==5)], c='k', s=50)
plt.plot(x, yp)
plt.ylim(-20, 100);
```

# $f(x) = x^3$

$\lim\limits_{h \to 0} x^3 = \frac{f(x+h)-f(x)}{h}$

$= \frac{(x+h)^3 - x^3}{h}$

$= \frac{(x+h)(x^2 + 2hx + h^2) - x^3}{h}$

$= \frac{(x^3 + 2hx^2 + xh^2 + hx^2 + 2xh^2 + h^3) - x^3}{h}$

$= \frac{3hx^2 + 3xh^2 + h^3}{h}$

$= 3x^2 + 3xh + h^2$

$h = 0$

$\Rightarrow \frac{dy}{dx} = 3x^2$

```python
# f(x) = x^3
# f'(x) = 3x^2

h = 1e-10
x = 2

print('Estimate:', ((x + h)**3 - x**3) / h)
print('True:', 3 * x**2)
print('\n')
print('Estimate:', ((x + h)**3 - (x - h)**3) / (2 * h))
print('True:', 3 * x**2)
```

# $f(x) = \frac{1}{x}$

$\lim\limits_{x \to 0} \frac{1}{x} = \frac{f(x+h) - f(x)}{h}$

$\lim\limits_{x \to 0} \frac{1}{x} = \frac{\frac{1}{x + h} - \frac{1}{x}}{h}$

$= \frac{1}{h}(\frac{1}{x+h} - \frac{1}{x})$

$= \frac{1}{h}(\frac{x}{x(x+h)} - \frac{x+h}{x(x+h)})$

$= \frac{1}{h}(\frac{x-(x+h)}{x(x+h)})$

$= \frac{1}{h}(\frac{x-x-h}{x(x+h)})$

$= \frac{1}{h}(\frac{-h}{x(x+h)})$

$= \frac{-1}{x(x+h)}$

$h = 0$

$\frac{dy}{dx} = -\frac{1}{x^2} \Rightarrow -x^{-2}$

```python
# f(x) = 1/x
# f'(x) = -1/x^2

h = 1e-10
x = 1

print('Estimate:', (1/(x+h) - 1/x) / h)
print('True:', -1/x**2)
print('\n')
print('Estimate:', (1/(x+h) - 1/(x - h)) / (2 * h))
print('True:', -1/x**2)
```

# $f(x) = \sqrt{x}$

$\lim\limits_{h \to 0} \sqrt{x} = \frac{f(x+h)_ - f(x)}{h}$

$= \frac{\sqrt{x+h} - \sqrt{x}}{h}$

$= \frac{\sqrt{x+h} - \sqrt{x}}{h} \times \frac{\sqrt{x+h} + \sqrt{x}}{\sqrt{x+h} + \sqrt{x}} \simeq (a+b)(a-b) = a^2 - b^2$

$= \frac{x + h - x}{h(\sqrt{x+h} + \sqrt{x})}$

$= \frac{h}{h(\sqrt{x+h} + \sqrt{x})}$

$= \frac{1}{\sqrt{x+h} + \sqrt{x}}$

$h = 0$

$\frac{1}{\sqrt{x+0} + \sqrt{x}}$

$\frac{1}{\sqrt{x} + \sqrt{x}}$

$\frac{1}{2\sqrt{x}}$

$= \frac{1}{2}x^{-\frac{1}{2}}$

```python
# f(x) = sqrt(x)
# f'(x) = 0.5 / sqrt(x)

h = 1e-10
x = 3

print('Estimate:', (np.sqrt(x+h) - np.sqrt(x)) / h)
print('True:', 0.5 / np.sqrt(x))
print('\n')
print('Estimate:', (np.sqrt(x+h) - np.sqrt(x-h)) / (2*h))
print('True:', 0.5 / np.sqrt(x))
```
