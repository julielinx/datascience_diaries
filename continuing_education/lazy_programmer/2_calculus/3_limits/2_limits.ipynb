# $\lim\limits_{x \to \infty} \frac{1}{x} = 0$

```python
x = np.linspace(-5, 5, 1000)
y = 1 / x

plt.plot(x, y)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.ylim(-15, 15);
```

# $\lim\limits_{x \to \infty} (1 + \frac{1}{x})^x = e$

# $\lim\limits_{x \to 0} \frac{\cos(x) - 1}{x} = 0$

```python
x = np.linspace(-50, 50, 1000)
y = (np.cos(x)-1)/x

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```
