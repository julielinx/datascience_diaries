```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-50, 50, 1000)
y = np.sin(x)/x
plt.plot(x, y)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

```python
plt.x
```

```python
y1 = 1/x
plt.plot(x, y1)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.xlim(-10, 10),
plt.ylim(-10, 10);
```

```python
y2 = x / (x -1)
plt.plot(x, y2)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.xlim(-10, 10),
plt.ylim(-10, 10);
```

```python
y3 = np.log2(x)
plt.plot(x, y3)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

```python
y4 = np.log10(x)
plt.plot(x, y4)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

```python
y5 = np.exp(x)
plt.plot(x, y5)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.xlim(-5, 5)
plt.ylim(0, 5);
```

```python
x1 = np.linspace(-5, 5, 1000)
y5 = np.e ** x1
plt.plot(x1, y5)
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.ylim(0, 10);
```
