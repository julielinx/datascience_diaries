## Dot product multiplication

```python
E = np.matmul(A.T, B)
E.shape
```

```python
pd.DataFrame(E)
```

```python
print(A.shape)
print(B.shape)
```

```python
def forloop_matx_mul(A, B):
    At = A.T
    C = []
    for ri in range(At.shape[0]):
        row = At[ri, :]
        mat_row = []
        for ci in range(B.shape[0]):
            col = B[:, ci]
            m_row = sum(row*col)
            mat_row.append(m_row)
        C.append(mat_row)
    return C
```

```python
pd.DataFrame(forloop_matx_mul(A, B))
```

```python
%%timeit
np.matmul(A.T, B)
```

```python
%%timeit
forloop_matx_mul(A, B)
```
