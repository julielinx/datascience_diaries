Circles formula:

radius 5:

- $x^2 + y^2 = 25$
- => $y^2 = 25 - x^2$
- => $y = \pm \sqrt{25 - x^2}$

radius 2.5

- $x^2 + y^2 = 6.25$
- => $y^2 = 6.25 - x^2$
- => $y = \pm \sqrt{6.25 - x^2}$

```python
x1 = np.linspace(-5, 5, 1000)
y1 = np.sqrt(abs(25 - x1**2))
y1 = np.concatenate((y1, y1 * -1))
x1 = np.concatenate((x1, x1))

x2 = np.linspace(-2.5, 2.5, 1000)
y2 = np.sqrt(abs(6.25 - x2**2))
y2 = np.concatenate((y2, y2 * -1))
x2 = np.concatenate((x2, x2))
```

```python
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k')
plt.plot(x1, y1)
plt.axis('scaled');
```
