# amazon-sagemaker-examples

## advanced_functionality

- Gluon: open source deep learning library created by AWS and Microsoft
- causal-inference: build your own container so you can package your own algorithm for training and deployment in SageMaker
- custom-training-containers: build your own container so you can package your own algorithm for training in SageMaker
- data_distribution_types: how to structure data to take advantage of distrubuted processing
- distributed_tensorflow_mask_rcnn: distributed TensorFlow training for TensorPack Mask/Faster-RCNN and AWS Samples Mask R-CNN
- fastai: deep learning library
