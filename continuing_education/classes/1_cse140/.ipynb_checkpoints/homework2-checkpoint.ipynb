```python
def load_dna_data(filename):
    inputfile = open(filename)
    linenum = 0
    seq = ""

    for line in inputfile:
        linenum = linenum + 1
        # if we are on the 2nd, 6th, 10th line...
        if linenum % 4 == 2:
            # Remove the newline characters from the end of the line
            line = line.rstrip()
            seq = seq + line
    return seq

def count_dna(data):
    g_ct = data.count('G')
    c_ct = data.count('C')
    a_ct = data.count('A')
    t_ct = data.count('T')
    gc_ct = g_ct + c_ct
    at_ct = a_ct + t_ct
    sum_ct = gc_ct + at_ct
    total_ct = len(data)
    gc_content = gc_ct / sum_ct
    at_content = at_ct / sum_ct
    at_gc_ratio = at_ct / gc_ct
    answer = f'''
    GC-content: {gc_content}
    AT-content: {at_content}
    G count: {g_ct}
    C count: {c_ct}
    A count: {a_ct}
    T count: {t_ct}
    Sum count: {sum_ct}
    Total count: {total_ct}
    AT/GC Ratio: {at_gc_ratio}'''
    print(answer)
    if gc_content >= 0.6:
        print(f'    GC Classification: high GC content')
    elif gc_content <= 0.4:
        print(f'    GC Classification: low GC content')
    else:
        print(f'    GC Classification: moderate GC content')
    
```
