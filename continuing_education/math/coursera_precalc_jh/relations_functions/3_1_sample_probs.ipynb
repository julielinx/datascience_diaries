## 1. Determine whether each function is increasing or decreasing

```python
import numpy as np
import matplotlib.pyplot as plt
```

a. $f(x) = 0.7^x$

Decreasing. At values of x < 0, the result will be large, extending toward $\infty$. At values of x > 0 values will continue to get smaller toward the horizontal asymptote of 0.

```python
x = np.linspace(-8, 8, 1000)
y = 0.7**x

plt.plot(x,y)
plt.grid()
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

b. $g(x) = 3^x$

This is a standard exponential function and will continue to increase toward $\infty$. Horizontal asymptote at 0.

```python
x = np.linspace(-5, 5, 100)
y = 3**x

plt.plot(x,y)
plt.grid()
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

c. $h(x) = 4^{-x}$

This is a mirrored exponential function across the y axis. It will continally decrease toward the y=0 asymptote.

It can also be thought of as a fraction: $h(x) = (\frac{1}{4})^x$

```python
x = np.linspace(-5, 5, 100)
y = 4**(-x)

plt.plot(x,y)
plt.grid()
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

d. $k(x) = (\frac{5}{6})^{-x} \simeq \frac{6}{5}^x$

This is a standard exponential function and will continue to increase.

```python
x = np.linspace(-15, 15, 1000)
y = (5/6)**(-x)

plt.plot(x,y)
plt.grid()
plt.axhline(y=0, c='k')
plt.axvline(x=0, c='k');
```

## 2. Find the domain and range of $f(x) = -3e^x + 2$

$e^x$ has a domain of $\mathbb{R}$ and a range of $(0, \infty)$

Multiplying this by a negative will flip it across the x-axis leaving the domain as $\mathbb{R}$, but changing the range to $(-\infty, 0)$

Finally, add 2 to this range: Domain = $\mathbb{R}$, Range = $(-\infty, 2)$
