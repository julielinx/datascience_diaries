## 2.1: Equations of Lines and Quadratics

### 1a. Find a line that passes through point (-2, 1) and is parallel to $y - 1 = 3(x + \frac{1}{2})$

$y - 1 = 3(x + \frac{1}{2}) \Rightarrow y = 3(x + \frac{1}{2}) + 1$ \
$y = 3x + 2\frac{1}{2}$\
$\Rightarrow \text{y-intercept} = 2\frac{1}{2}, \text{slope} = 3$

For point (-2, 1): \
$\quad 1 = 3(-2) + u$ \
$u = 7$

Solution: $y = 3x + 7$

```python
x = np.linspace(-10, 10, 1000)
y1 = 3*x + 2.5
y2 = 3*x + 7

plt.plot(x, y1, label='3x + 2.5')
plt.plot(x, y2, label='3x + 7')
plt.scatter(-2, 1)
plt.grid()
plt.legend();
```

### 1b. Find a line that passes through point (-2, 1) and is perpendicular to the line $4y = 1 - x$

$4y = 1 - x \Rightarrow y = \frac{1 - x}{4} = -\frac{1}{4}x + \frac{1}{4}$\
$\text{y-intercept} = \frac{1}{4}, \text{slope} = -\frac{1}{4}$

Perpendicular slope = $m_1(m_2) = -1$\
$m_2 = \frac{-1}{m_1}$
$m_2 = \frac{-1}{-\frac{1}{4}}$\
$m_2 = -4(-1) = 4$

$1 = 4(-2) + u$\
$u = 9$

Solution: $y = 4x +9$

```python
x = np.linspace(-10, 10, 100)
y1 = (-(1/4) * x) + (1/4)
y2 = (4*x) + 9

plt.plot(x, y1, label='-1/4x + 1/4')
plt.plot(x, y2, label='4x+9')
plt.scatter(-2, 1)
plt.grid()
plt.legend()
plt.axis('equal');
```
